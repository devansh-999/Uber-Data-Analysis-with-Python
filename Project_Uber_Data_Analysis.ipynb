**Importing the necessary libraries**

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns

import datetime
import calendar
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/uberdrive.csv')
data.head()
```

```python
data.isnull()
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'],format='%m/%d/%Y %H:%M')
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'],format='%m/%d/%Y %H:%M')
```

```python
data.dtypes
```

```python
hour = []
day = []
dayofweek = []
month = []
weekday = []

for x in data['START_DATE*']:
  hour.append(x.hour)
  day.append(x.day)
  dayofweek.append(x.dayofweek)
  month.append(x.month)
  weekday.append(calendar.day_name[dayofweek[-1]])

data['HOUR'] = hour
data['DAY'] = day
data['DAY OF WEEK'] = dayofweek
data['MONTH'] = month
data['WEEKDAY'] = weekday
```

```python
data.head()
```

#**Data Visualisation**

```python
data['CATEGORY*'].value_counts()
```

**What categories do we have**

```python
sns.countplot(x='CATEGORY*',data = data)
```

**How long do people travel with Uber**

```python
data['MILES*'].plot.hist()
```

**What Hour Do Most People Take Uber To Their Destination?**

```python
hours = data['START_DATE*'].dt.hour.value_counts()
hours.plot(kind='bar',color = 'pink',figsize = (10,5))

plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Number of Trips vs Hours')
```

**Check The Purpose Of Trips**

```python
data['PURPOSE*'].value_counts().plot(kind='bar',color = 'purple',figsize = (10,5))
plt.xlabel('Purpose')
plt.ylabel('Frequency')
plt.title('Purpose of Trips with its frequency')
```

**Which Day Has The Highest Number Of Trips**

```python
data['WEEKDAY'].value_counts().plot(kind='bar',color = 'grey',figsize = (10,5))

plt.xlabel('Day')
plt.ylabel('Frequency')
plt.title('Number of Trips vs Day')
```

**What Are The Number Of Trips Per Each Day?**

```python
data['DAY'].value_counts().plot(kind='bar',color = 'red',figsize = (10,5))

plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Number of Trips vs Day')
```

**What Are The Trips In The Month**

```python
data['MONTH'].value_counts().plot(kind='bar',color = 'blue',figsize = (10,5))

plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Number of Trips vs Month')
```

**Where Do People Start Boarding Their Trip From Most?**

```python
data['START*'].value_counts().plot(kind='bar',color = 'yellow',figsize = (30,8))
```
